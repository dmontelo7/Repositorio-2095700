```python
# Parte 1: Carga y Limpieza (Pandas y NumPy)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Leer el archivo .csv de la URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

# 2. Imprimir las primeras 5 filas y mostrar los tipos de datos
print("--- Primeras 5 filas ---")
print(df.head())
print("\n--- Tipos de datos por columna ---")
print(df.dtypes)

# 3. Identificar los valores nulos en horsepower
nulos_antes = df['horsepower'].isnull().sum()
print(f"\nValores nulos en horsepower antes: {nulos_antes}")

# Reemplazar nulos con la mediana usando Pandas/NumPy
mediana_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(mediana_hp)

nulos_despues = df['horsepower'].isnull().sum()
print(f"Valores nulos en horsepower después: {nulos_despues} (reemplazados con la mediana: {mediana_hp})")

# 4. Crear la columna relacion_peso_potencia
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']

print("\n--- Relación peso-potencia ---")
print(df[['weight', 'horsepower', 'relacion_peso_potencia']].head())
```

```python
# Parte 2: Visualización de Datos (Matplotlib)
```

```python
plt.style.use('ggplot')

# 1. Histograma de aceleración
plt.figure(figsize=(8, 5))
plt.hist(df['acceleration'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribución de la Aceleración de los Vehículos')
plt.xlabel('Aceleración')
plt.ylabel('Frecuencia')
plt.show()

# 2. Gráfico de dispersión
plt.figure(figsize=(8, 5))
plt.scatter(df['displacement'], df['horsepower'], alpha=0.6, color='green')
plt.title('Desplazamiento del motor vs. Caballos de fuerza')
plt.xlabel('Desplazamiento')
plt.ylabel('Horsepower')
plt.show()

# 3. Gráfico de Barras
# Agrupar datos por país de origen y calcular el promedio de MPG
mpg_por_origen = df.groupby('origin')['mpg'].mean()

plt.figure(figsize=(8, 5))
mpg_por_origen.plot(kind='bar', color=['tomato', 'gold', 'lightgreen'])
plt.title('Promedio de MPG por país de origen')
plt.xlabel('País de origen')
plt.ylabel('Promedio de millas por galón (MPG)')
plt.xticks(rotation=0) # Mantener nombres de países horizontales
plt.show()
```
